# Matplotlib
다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

# 1. 그래프 기본

```python
x = [1, 2, 3]
y = [2, 4, 8]
plt.plot(x, y)
```

```python
plt.plot(x,y)
plt.show() # 텍스트 보기 싫을 때
```

## Title 설정

```python
plt.plot(x, y)
plt.title('Line Graph')
```

## 한글 폰트 설정

```python
import matplotlib.font_manager as fm
fm.fontManager.ttflist # 사용 가능한 폰트 확인
[f.name for f in fm.fontManager.ttflist]
```

```python
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.plot(x, y)
plt.title('꺾은선 그래프')
```

```python
plt.plot([-1, 0, 1], [-5, -1, 2])
```

# 2. 축

```python
plt.plot(x,y)
plt.title('꺾은선 그래프', fontdict={'family' : 'HYGungSo-Bold', 'size':20})
```

```python
plt.plot(x, y)
plt.xlabel('X축')
plt.ylabel('Y축')
```

```python
plt.plot(x, y)
plt.xlabel('X축', color = 'red')
plt.ylabel('Y축', color = '#00aa00') # 색상 RGB
```

```python
plt.plot(x, y)
plt.xlabel('X축', color = 'red', loc='right') # left, center, right
plt.ylabel('Y축', color = '#00aa00', loc='top') # top, center, bottom
```

```python
plt.plot(x,y)
plt.xticks([1,2,3])
plt.yticks([3,6,9,12])
plt.show()
```

# 3. 범례 (legend)

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend()
```

```python
plt.plot(x, y, label='무슨 데이터')
plt.legend(loc='upper center') # 가장 좋은 자리 = best
```

```python
plt.plot(x, y, label='범례')
plt.legend(loc=(0.5, 0.5)) # x축, y축 (0~1 사이)
```
