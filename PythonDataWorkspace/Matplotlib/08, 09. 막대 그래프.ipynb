# 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
plt.bar(labels, values, color='r')
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
colors = ['r', 'g', 'b']
plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
plt.bar(labels, values)
plt.ylim(175, 195)
```

```python
plt.bar(labels, values, width= 0.5)
plt.xticks(rotation=45) # x 축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45) # y 축의 이름 데이터 각도를 45도로 설정
```

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
ticks = ['1번학생', '2번학생', '3번학생']
plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=90)
```

# 9. 막대 그래프 (심화)

```python
labels = ['강백호', '서태웅', '정대만']
values = [190, 187, 184]
plt.barh(labels, values)
plt.xlim(175, 195)
```

```python
bar = plt.bar(labels, values)
bar[0].set_hatch('/') # ////
bar[1].set_hatch('x') # xxxx
bar[2].set_hatch('..') # ...
```

```python
bar = plt.bar(labels, values)
plt.ylim(175, 195)

for idx, rect in enumerate(bar):
    plt.text(idx, rect.get_height()+0.5, values[idx], ha='center', color='blue')
```
